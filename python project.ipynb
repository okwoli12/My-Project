```python
def mean(col):
    b = df[col].mean()
    df[col] = df[col].fillna(b)
mean("2003")
mean("2004")
mean("2005")
mean("2006")
mean("2007")
mean("2008")
mean("2009")
mean("2010")
mean("2011")
mean("2012")
mean("2013")
mean("2014")
mean("2015")
mean("2016")
mean("2017")
mean("2018")
mean("2019")
mean("2020")
```

```python
row_sums = []
for i in range(len(df)):
    row_sum = sum(df.iloc[i, 5:]) # Assumes the first column is an index and should be excluded
    row_sums.append(row_sum)
row_sums
df["row_sums"] = row_sums
df
```

```python
df.isna().sum()
```

```python
total = df.groupby ('Area')["row_sums"].sum().sort_values(ascending = False)
total1=total.reset_index()
```
